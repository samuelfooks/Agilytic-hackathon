```python
import os
import geopandas as gpd
import pandas

filelist = []

for (root, directory, files) in os.walk("data/cycle_networks/"):
    for x in files:
        if x.endswith(".gpkg"):
            filelist.append(os.path.join(root,x))
```

```python
df = gpd.GeoDataFrame()
for file in filelist:
    tmp = gpd.read_file(file)
    df = pandas.concat([df,tmp])
```

```python
df.to_file("data/cycle_network_all",driver="GPKG")
```
